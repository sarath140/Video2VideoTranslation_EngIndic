```python
from openai import OpenAI
from google.colab import drive, userdata
import os
import glob

# === Step 1: Mount Google Drive ===
drive.mount('/content/drive')
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')


# === Step 2: Set up directories ===
base_dir = "/content/drive/My Drive/OpenAI_API_pipeline"
input_dir = os.path.join(base_dir, "input")   # folder containing your audio/video files
asr_dir = os.path.join(base_dir, "asr")       # folder to save transcriptions

os.makedirs(input_dir, exist_ok=True)
os.makedirs(asr_dir, exist_ok=True)

print("‚úÖ Folders ready:")
print("Input folder:", input_dir)
print("ASR output folder:", asr_dir)

# === Step 3: Initialize OpenAI client ===
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

# === Step 4: Process all media files in input folder ===
supported_exts = (".mp3", ".mp4", ".wav", ".m4a", ".mov", ".mkv")
files = [f for f in glob.glob(os.path.join(input_dir, "*")) if f.lower().endswith(supported_exts)]

if not files:
    print("No audio/video files found in:", input_dir)
else:
    print(f"Found {len(files)} file(s) for transcription.\n")

for file_path in files:
    base = os.path.splitext(os.path.basename(file_path))[0]
    txt_path = os.path.join(asr_dir, f"{base}.txt")
    srt_path = os.path.join(asr_dir, f"{base}.srt")

    print(f"üîÑ Processing: {os.path.basename(file_path)}")

    # --- Generate TXT ---
    with open(file_path, "rb") as f:
        txt = client.audio.transcriptions.create(
            model="whisper-1",
            file=f,
            language="hi",
            response_format="text"
        )

    # --- Generate SRT ---
    with open(file_path, "rb") as f:
        srt = client.audio.transcriptions.create(
            model="whisper-1",
            file=f,
            response_format="srt"
        )

    # --- Save outputs in Drive ---
    with open(txt_path, "w", encoding="utf-8") as f:
        f.write(txt)
    with open(srt_path, "w", encoding="utf-8") as f:
        f.write(srt)

    print(f"‚úÖ Saved: {txt_path}")
    print(f"‚úÖ Saved: {srt_path}\n")

print("üéâ All files processed successfully!")
```

# MT multi file srt batch

```python
from openai import OpenAI
from google.colab import drive, userdata
import os
import re
import time

# === Mount Google Drive and API ===
drive.mount('/content/drive')
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

# === Paths ===
base_dir = "/content/drive/My Drive/OpenAI_API_pipeline"
asr_dir = os.path.join(base_dir, "asr")  # input SRTs
mt_dir = os.path.join(base_dir, "mt")   # translated output
os.makedirs(mt_dir, exist_ok=True)

target_language = "Telugu"
print("üü¢ Ready ‚Äî Processing all .srt files...")

# === SRT parsing pattern ===
pattern = r"(\d+)\s+([\d:,]+ --> [\d:,]+)\s+(.+?)(?=\n\d+\n|$)"

def translate_batch(lines):
    """Translate list of subtitle text chunks at once."""
    joined_text = "\n".join(lines)
    prompt = f"""
Translate this subtitle dialog into {target_language}.
Preserve meaning. Keep subtitles short & natural.
Do NOT translate numbers or timestamps.

Text:
{joined_text}
"""
    for _ in range(3):  # retry logic
        try:
            response = client.chat.completions.create(
                model="gpt-5",
                messages=[
                    {"role": "system", "content": "You are a professional subtitle translator for Indic languages."},
                    {"role": "user", "content": prompt}
                ],
            )
            return response.choices[0].message.content.split("\n")
        except Exception as e:
            print("Retrying batch due to error:", e)
            time.sleep(3)
    return [""] * len(lines)  # fallback

# === Loop over all SRT files ===
for f_name in os.listdir(asr_dir):
    if not f_name.lower().endswith(".srt"):
        continue

    input_file = os.path.join(asr_dir, f_name)
    print(f"\nüé¨ Processing: {f_name}")

    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()

    entries = re.findall(pattern, content, flags=re.DOTALL)
    print(f"   ‚Üí {len(entries)} subtitles detected")

    translated_entries = []
    translated_text_only = []

    batch_size = 15
    for i in range(0, len(entries), batch_size):
        batch = entries[i:i+batch_size]
        orig_texts = [t[2].strip() for t in batch]

        translated_batch = translate_batch(orig_texts)

        for (num, ts, _), trans in zip(batch, translated_batch):
            translated_entries.append(f"{num}\n{ts}\n{trans}\n")
            translated_text_only.append(trans)

        print(f"   ‚úÖ Translated segments {i+1}‚Äì{min(i+batch_size,len(entries))}")

    # Save outputs
    base = os.path.splitext(f_name)[0]
    srt_out = os.path.join(mt_dir, f"{base}_{target_language}.srt")
    txt_out = os.path.join(mt_dir, f"{base}_{target_language}.txt")

    with open(srt_out, "w", encoding='utf-8') as f:
        f.write("\n".join(translated_entries))

    with open(txt_out, "w", encoding='utf-8') as f:
        f.write("\n".join(translated_text_only))

    print(f"   üìÅ Saved ‚Üí {srt_out}")
    print(f"   üìÑ Saved ‚Üí {txt_out}")

print("\n‚úÖ All files translated successfully!")
```

```python
from openai import OpenAI
from google.colab import drive, userdata
import os
import re
import time

# === Mount Google Drive and API ===
drive.mount('/content/drive')
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

# === Paths ===
base_dir = "/content/drive/My Drive/tTest_28_Adnew_mp3/Test_28_Whisper1_asr/Hindi/"
asr_dir = os.path.join(base_dir, "Srt")  # input SRTs
mt_dir = os.path.join(base_dir, "mtnew")   # translated output
os.makedirs(mt_dir, exist_ok=True)

target_language = "English"
print("üü¢ Ready ‚Äî Processing all .srt files...")

# === SRT parsing pattern ===
pattern = r"(\d+)\s+([\d:,]+ --> [\d:,]+)\s+(.+?)(?=\n\d+\n|$)"

def translate_batch(lines):
    """Translate list of subtitle text chunks with alignment markers."""
    # Add numbered markers so GPT knows what belongs together
    numbered_text = "\n".join([f"{i+1}. {line}" for i, line in enumerate(lines)])
    prompt = f"""
Translate each subtitle line into {target_language}.
Return the result in the same numbered list format.
Do NOT merge or skip lines.
Do NOT translate numbers, timestamps, or markers.

Example:
Input:
1. ‡§®‡§Æ‡§∏‡•ç‡§§‡•á
2. ‡§ï‡•à‡§∏‡•á ‡§π‡•ã?

Output:
1. Hello
2. How are you?

Now translate:
{numbered_text}
"""

    for _ in range(3):
        try:
            response = client.chat.completions.create(
                model="gpt-5",
                messages=[
                    {"role": "system", "content": "You are a professional subtitle translator for Indic languages."},
                    {"role": "user", "content": prompt}
                ],
            )
            raw_output = response.choices[0].message.content

            # Extract translations by matching numbered lines
            matches = re.findall(r'^\s*\d+\.\s*(.+)', raw_output, re.MULTILINE)
            if len(matches) != len(lines):
                print(f"‚ö†Ô∏è Warning: expected {len(lines)} lines, got {len(matches)} ‚Äî auto-filling blanks")
                # Pad or trim as needed
                while len(matches) < len(lines):
                    matches.append("")
                matches = matches[:len(lines)]

            return matches
        except Exception as e:
            print("Retrying batch due to error:", e)
            time.sleep(3)
    return [""] * len(lines)

# === Loop over all SRT files ===
for f_name in os.listdir(asr_dir):
    if not f_name.lower().endswith(".srt"):
        continue

    input_file = os.path.join(asr_dir, f_name)
    print(f"\nüé¨ Processing: {f_name}")

    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()

    entries = re.findall(pattern, content, flags=re.DOTALL)
    print(f"   ‚Üí {len(entries)} subtitles detected")

    translated_entries = []
    translated_text_only = []

    batch_size = 10
    for i in range(0, len(entries), batch_size):
        batch = entries[i:i+batch_size]
        orig_texts = [t[2].strip() for t in batch]

        translated_batch = translate_batch(orig_texts)

        for (num, ts, _), trans in zip(batch, translated_batch):
            translated_entries.append(f"{num}\n{ts}\n{trans}\n")
            translated_text_only.append(trans)

        print(f"   ‚úÖ Translated segments {i+1}‚Äì{min(i+batch_size,len(entries))}")

    # Save outputs
    base = os.path.splitext(f_name)[0]
    srt_out = os.path.join(mt_dir, f"{base}_{target_language}.srt")
    txt_out = os.path.join(mt_dir, f"{base}_{target_language}.txt")

    with open(srt_out, "w", encoding='utf-8') as f:
        f.write("\n".join(translated_entries))

    with open(txt_out, "w", encoding='utf-8') as f:
        f.write("\n".join(translated_text_only))

    print(f"   üìÅ Saved ‚Üí {srt_out}")
    print(f"   üìÑ Saved ‚Üí {txt_out}")

print("\n‚úÖ All files translated successfully!")
```

# tts 40mini no token limit chunked

```python
from openai import OpenAI
from google.colab import drive, userdata
from IPython.display import Audio
from pydub import AudioSegment
from tqdm import tqdm
import os, shutil, textwrap, time

# === Mount Google Drive ===
drive.mount('/content/drive', force_remount=True)

# === Setup OpenAI client ===
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# === Input & Output paths ===
file_path = "/content/drive/MyDrive/OpenAI_API_pipeline/mt/multi srt in and sam lang srt txt out(telugu run 2)/Chapter 5B - The Concept of Book Keeping and its Fundamental_Telugu.txt"
drive_folder = "/content/drive/MyDrive/OpenAI_API_pipeline/tts/"
os.makedirs(drive_folder, exist_ok=True)

# === Read text ===
with open(file_path, "r", encoding="utf-8") as f:
    text_content = f.read().strip()

print(f"üìñ Loaded {len(text_content)} characters from file")

# === Split into smaller chunks (‚âà4000 chars each) ===
chunks = textwrap.wrap(text_content, width=4000, break_long_words=False, replace_whitespace=False)
print(f"‚úÇÔ∏è Split into {len(chunks)} chunks (to fit model limits)")

# === Generate TTS for each chunk ===
tts_parts = []
for i, chunk in enumerate(tqdm(chunks, desc="Generating speech"), start=1):
    out_path = f"tts_part_{i}.mp3"
    success = False

    for attempt in range(3):  # Retry up to 3 times per chunk
        try:
            with client.audio.speech.with_streaming_response.create(
                model="gpt-4o-mini-tts",
                voice="alloy",  # Try "soft" or "verse" for variation
                input=chunk
            ) as response:
                response.stream_to_file(out_path)
            print(f"‚úÖ Chunk {i} saved ({len(chunk)} chars)")
            tts_parts.append(AudioSegment.from_mp3(out_path))
            success = True
            break
        except Exception as e:
            print(f"‚ö†Ô∏è Error in chunk {i}, attempt {attempt+1}: {e}")
            time.sleep(2)
            # If still too long, reduce chunk size and retry
            if "maximum input limit" in str(e) and len(chunk) > 1000:
                print(f"‚úÇÔ∏è Reducing chunk {i} size and retrying...")
                chunk = chunk[:int(len(chunk) * 0.7)]
            else:
                break

    if not success:
        print(f"‚ùå Failed to process chunk {i}, skipping.")

# === Merge all parts ===
if tts_parts:
    final_audio = sum(tts_parts)
    final_path = os.path.join(drive_folder, "Chapter_5B_Telugu_TTS_Combined.mp3")
    final_audio.export(final_path, format="mp3")
    print(f"\nüîä Combined audio saved ‚Üí {final_path}")
else:
    final_path = None
    print("\n‚ùå No chunks were successfully converted to audio.")

# === Play in Colab ===
if final_path:
    Audio(final_path)
```

#tts using duration match (en srt)

```python
!pip install --upgrade openai pysrt tqdm pydub

from openai import OpenAI
from google.colab import drive, userdata
from IPython.display import Audio
from pydub import AudioSegment
from tqdm import tqdm
import pysrt, os, shutil, time

# -------------------------------
# Setup
# -------------------------------
drive.mount('/content/drive', force_remount=True)
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# === Path to your .srt file ===
srt_file_path = "/content/drive/MyDrive/aa/test2.srt"

# === Output folder ===
base_name = os.path.splitext(os.path.basename(srt_file_path))[0]
output_dir = f"/content/drive/MyDrive/aaOai/{base_name}_segments_openai"
os.makedirs(output_dir, exist_ok=True)

# -------------------------------
# Helper: parse SRT
# -------------------------------
def parse_srt(path):
    subs = pysrt.open(path)
    segments = []
    for sub in subs:
        start = sub.start.hours*3600 + sub.start.minutes*60 + sub.start.seconds + sub.start.milliseconds/1000
        end   = sub.end.hours*3600   + sub.end.minutes*60   + sub.end.seconds   + sub.end.milliseconds/1000
        text = sub.text.replace("\n", " ").strip()
        if text:
            segments.append((start, end, text))
    return segments

segments = parse_srt(srt_file_path)
print(f"üé¨ Found {len(segments)} subtitle segments")

# -------------------------------
# Helper: generate TTS with retry
# -------------------------------
def tts_with_retry(text, voice="alloy", retries=3, delay=3):
    for attempt in range(retries):
        try:
            with client.audio.speech.with_streaming_response.create(
                model="gpt-4o-mini-tts",
                voice=voice,
                input=text
            ) as response:
                temp_file = f"temp_{int(time.time())}.mp3"
                response.stream_to_file(temp_file)
            return temp_file
        except Exception as e:
            print(f"‚ö†Ô∏è Error on attempt {attempt+1}: {e}")
            if "maximum input limit" in str(e) and len(text) > 500:
                text = text[: int(len(text) * 0.7)]  # shorten
                print("‚úÇÔ∏è Reduced text length and retrying...")
            time.sleep(delay)
    return None

# -------------------------------
# Helper: match segment duration
# -------------------------------
def match_duration(audio_path, target_duration):
    """Ensure TTS matches target segment duration by adding silence or adjusting speed."""
    audio = AudioSegment.from_file(audio_path)
    current_dur = len(audio) / 1000.0  # seconds

    if abs(current_dur - target_duration) < 0.05:
        return audio  # close enough

    if current_dur < target_duration:
        # Add silence at end to reach target duration
        silence = AudioSegment.silent(duration=(target_duration - current_dur) * 1000)
        return audio + silence
    else:
        # Slightly speed up if audio is too long (max 15%)
        ratio = target_duration / current_dur
        if ratio < 0.85:
            ratio = 0.85  # prevent extreme speed-up
        audio = audio._spawn(audio.raw_data, overrides={
            "frame_rate": int(audio.frame_rate * ratio)
        }).set_frame_rate(audio.frame_rate)
        # Add silence padding if needed after stretch
        new_dur = len(audio) / 1000.0
        if new_dur < target_duration:
            silence = AudioSegment.silent(duration=(target_duration - new_dur) * 1000)
            audio += silence
        return audio

# -------------------------------
# Main loop: generate & align
# -------------------------------
failed_log = os.path.join(output_dir, "failed_segments.txt")
tts_parts = []

for idx, (start, end, text) in enumerate(tqdm(segments, desc="Generating TTS"), start=1):
    if len(text.strip()) < 3:
        continue

    target_duration = end - start
    print(f"\nSegment {idx}: {start:.2f}s ‚Üí {end:.2f}s | {target_duration:.2f}s | {text[:60]}...")

    temp_file = tts_with_retry(text, voice="alloy")

    if not temp_file:
        print(f"‚ùå Failed segment {idx}")
        with open(failed_log, "a", encoding="utf-8") as log:
            log.write(f"{idx}: {text}\n")
        continue

    # Match duration
    matched_audio = match_duration(temp_file, target_duration)

    # Save
    final_path = os.path.join(output_dir, f"{idx:03d}.mp3")
    matched_audio.export(final_path, format="mp3")
    os.remove(temp_file)
    tts_parts.append((final_path, start, end))
    print(f"‚úÖ Duration matched & saved ‚Üí {final_path}")

# -------------------------------
# Combine all with silence gaps
# -------------------------------
if tts_parts:
    print("\nüîä Combining all segments (with silence gaps)...")
    combined = AudioSegment.silent(duration=0)
    last_end = 0

    for fpath, start, end in tts_parts:
        audio = AudioSegment.from_mp3(fpath)
        if start > last_end:
            # Add silence between last and current start
            silence_gap = AudioSegment.silent(duration=(start - last_end) * 1000)
            combined += silence_gap
        combined += audio
        last_end = end

    combined_out = os.path.join(output_dir, f"{base_name}_timed_combined.mp3")
    combined.export(combined_out, format="mp3")
    print(f"‚úÖ Combined timed audio saved ‚Üí {combined_out}")
    display(Audio(combined_out))
else:
    print("‚ùå No segments converted successfully.")

print(f"\nüìÇ All files saved in: {output_dir}")
```
