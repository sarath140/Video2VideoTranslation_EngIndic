```python
from openai import OpenAI
from google.colab import drive, userdata
from IPython.display import Audio
from pydub import AudioSegment
from tqdm import tqdm
import os, shutil, textwrap, time

# === Mount Google Drive ===
drive.mount('/content/drive', force_remount=True)

# === Setup OpenAI client ===
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# === Input & Output paths ===
file_path = "/content/drive/MyDrive/TTS_text_files_test/2A/Mod_chap2A_eng.txt"
drive_folder = "/content/drive/MyDrive/TTS_text_files_test/2A/TTS"
os.makedirs(drive_folder, exist_ok=True)

# === Read text ===
with open(file_path, "r", encoding="utf-8") as f:
    text_content = f.read().strip()

print(f"ğŸ“– Loaded {len(text_content)} characters from file")

# === Split into smaller chunks (â‰ˆ4000 chars each) ===
chunks = textwrap.wrap(text_content, width=4000, break_long_words=False, replace_whitespace=False)
print(f"âœ‚ï¸ Split into {len(chunks)} chunks (to fit model limits)")

# === Generate TTS for each chunk ===
tts_parts = []
for i, chunk in enumerate(tqdm(chunks, desc="Generating speech"), start=1):
    out_path = f"tts_part_{i}.mp3"
    success = False

    for attempt in range(3):  # Retry up to 3 times per chunk
        try:
            with client.audio.speech.with_streaming_response.create(
                model="gpt-4o-mini-tts",
                voice="alloy",  # Try "soft" or "verse" for variation
                input=chunk
            ) as response:
                response.stream_to_file(out_path)
            print(f"âœ… Chunk {i} saved ({len(chunk)} chars)")
            tts_parts.append(AudioSegment.from_mp3(out_path))
            success = True
            break
        except Exception as e:
            print(f"âš ï¸ Error in chunk {i}, attempt {attempt+1}: {e}")
            time.sleep(2)
            # If still too long, reduce chunk size and retry
            if "maximum input limit" in str(e) and len(chunk) > 1000:
                print(f"âœ‚ï¸ Reducing chunk {i} size and retrying...")
                chunk = chunk[:int(len(chunk) * 0.7)]
            else:
                break

    if not success:
        print(f"âŒ Failed to process chunk {i}, skipping.")

# === Merge all parts ===
if tts_parts:
    final_audio = sum(tts_parts)
    final_path = os.path.join(drive_folder, "chap1a_mal.mp3")
    final_audio.export(final_path, format="mp3")
    print(f"\nğŸ”Š Combined audio saved â†’ {final_path}")
else:
    final_path = None
    print("\nâŒ No chunks were successfully converted to audio.")

# === Play in Colab ===
if final_path:
    Audio(final_path)
```

# diff voices for a single telugu file

```python
from openai import OpenAI
from google.colab import drive, userdata
from IPython.display import Audio
from pydub import AudioSegment
from tqdm import tqdm
import os, shutil, textwrap, time

# === Mount Google Drive ===
drive.mount('/content/drive', force_remount=True)

# === Setup OpenAI client ===
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# === Input & Output paths ===
file_path = "/content/drive/MyDrive/TTS_text_files_test/1A Eng/chap1a_tel.txt"
drive_folder = "/content/drive/MyDrive/TTS_text_files_test/1A Eng/TTS/DiffVoice"
os.makedirs(drive_folder, exist_ok=True)

# === Read text ===
with open(file_path, "r", encoding="utf-8") as f:
    text_content = f.read().strip()

print(f"ğŸ“– Loaded {len(text_content)} characters from file")

# === Split into smaller chunks (â‰ˆ4000 chars each) ===
chunks = textwrap.wrap(text_content, width=4000, break_long_words=False, replace_whitespace=False)
print(f"âœ‚ï¸ Split into {len(chunks)} chunks (to fit model limits)")

# List of available voices
voices = ["alloy", "ash", "ballad", "coral", "echo", "fable", "nova", "onyx", "sage", "shimmer"]

# === Generate TTS for each chunk with all voices ===
for voice in voices:
    tts_parts = []
    print(f"\nğŸ¤ Testing with voice: {voice}")

    for i, chunk in enumerate(tqdm(chunks, desc=f"Generating speech with {voice}"), start=1):
        out_path = f"tts_{voice}_part_{i}.mp3"
        success = False

        for attempt in range(3):  # Retry up to 3 times per chunk
            try:
                with client.audio.speech.with_streaming_response.create(
                    model="gpt-4o-mini-tts",
                    voice=voice,  # Use the current voice in the loop
                    input=chunk
                ) as response:
                    response.stream_to_file(out_path)
                print(f"âœ… Chunk {i} saved ({len(chunk)} chars) with {voice}")
                tts_parts.append(AudioSegment.from_mp3(out_path))
                success = True
                break
            except Exception as e:
                print(f"âš ï¸ Error in chunk {i}, attempt {attempt+1}: {e}")
                time.sleep(2)
                # If still too long, reduce chunk size and retry
                if "maximum input limit" in str(e) and len(chunk) > 1000:
                    print(f"âœ‚ï¸ Reducing chunk {i} size and retrying...")
                    chunk = chunk[:int(len(chunk) * 0.7)]
                else:
                    break

        if not success:
            print(f"âŒ Failed to process chunk {i} for voice {voice}, skipping.")

    # === Merge all parts for the current voice ===
    if tts_parts:
        final_audio = sum(tts_parts)
        final_path = os.path.join(drive_folder, f"chap1a_{voice}_tel.mp3")
        final_audio.export(final_path, format="mp3")
        print(f"\nğŸ”Š Combined audio for {voice} saved â†’ {final_path}")
    else:
        print(f"\nâŒ No chunks were successfully converted to audio for {voice}.")

    # === Optional: Play in Colab ===
    Audio(final_path)  # Uncomment if you want to listen directly in Colab

```
