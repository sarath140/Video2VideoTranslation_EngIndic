```python
# ======================================================
# üîä Replace Audio in an MP4 (Files already in Google Drive)
# ======================================================

from google.colab import drive
import subprocess
import os

# --------------------------
# 1Ô∏è‚É£ Mount Google Drive
# --------------------------
drive.mount('/content/drive')

print("\n‚úÖ Google Drive mounted!")

# --------------------------
# 2Ô∏è‚É£ Choose input files
# --------------------------
# UPDATE THESE TWO PATHS

mp4_path = "/content/drive/MyDrive/Test1/Chapter 1A - Introduction to DTP.mp4"
wav_path = "/content/drive/MyDrive/Test1/tts/1A_merged_timed_Eng.wav"

# Example:
# mp4_path = "/content/drive/MyDrive/tts_tel_test/video.mp4"
# wav_path = "/content/drive/MyDrive/tts_tel_test/Chapter 2A - Green House..._merged_timed_Tamil.wav"

# --------------------------
# 3Ô∏è‚É£ Output path
# --------------------------
output_path = "/content/drive/MyDrive/Test1/tts/1A_eng_audio.mp4"


# --------------------------
# 4Ô∏è‚É£ Run FFmpeg to replace audio
# --------------------------
print(f"\nüé¨ Replacing audio:\nVideo: {mp4_path}\nAudio: {wav_path}\n")

cmd = [
    "ffmpeg", "-y",
    "-i", mp4_path,    # input video
    "-i", wav_path,    # input audio
    "-map", "0:v",     # keep original video
    "-map", "1:a",     # replace audio
    "-c:v", "copy",    # do NOT re-encode video
    "-c:a", "aac",     # MP4 requires AAC audio
    "-shortest",       # sync durations
    output_path
]

subprocess.run(cmd, check=True)

print("\n‚úÖ DONE! Your final video is saved at:\n", output_path)
```
