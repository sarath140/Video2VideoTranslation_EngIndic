```python
from huggingface_hub import hf_hub_download
from transformers import AutoProcessor, VocosModel

ckpt = hf_hub_download('charactr/vocos-mel-24khz', 'model.safetensors')
processor = AutoProcessor.from_pretrained('charactr/vocos-mel-24khz')
vocos = VocosModel.from_pretrained('charactr/vocos-mel-24khz')

print('Vocos ready!')
```

```python
from transformers import AutoModel

model = AutoModel.from_pretrained('ai4bharat/IndicF5', trust_remote_code=True)
print('IndicF5 loaded!')
```

```python
import soundfile as sf
import numpy as np

input_text = 'இது ஒரு சோதனை உரை.'
ref_audio = '/content/ref.wav'  # upload
ref_text = 'This is reference audio text.'

audio = model(input_text, ref_audio_path=ref_audio, ref_text=ref_text)
sf.write('out.wav', audio.astype(np.float32), 24000)
print('Saved out.wav')
```
